```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#设置黑体为默认字体
plt.rcParams["font.sans-serif"]="SimHei"
#设置正常显示负号
plt.rcParams["axes.unicode_minus"]=False
n=30
y1=np.random.randint(0,30,n)
y2=np.random.randint(50,80,n)
```

```python
plt.plot(y1,color="blue",label="温度")
plt.plot(y2,color="green",label="湿度")
plt.legend()
plt.xlim(0,23)
plt.ylim(0,80)
plt.xlabel("小时",fontsize=20)
plt.ylabel("测量值",fontsize=20)
plt.title("24小时湿度温度统计",fontsize=25)
```
