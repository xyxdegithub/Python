```python
import tensorflow as tf
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("image/Lena.jpg")
```

```python
img_r, img_g, img_b = img.split()
```

```python
plt.figure(figsize=(16, 16))

plt.subplot(151)
plt.axis("off")
plt.imshow(img)
plt.title("img", fontsize=20)

plt.subplot(152)

plt.imshow(img_r)
plt.title("img_r", fontsize=20)

plt.subplot(153)

plt.imshow(img_g)
plt.title("img_g", fontsize=20)

plt.subplot(154)

plt.imshow(img_b)
plt.title("img_b", fontsize=20)

img_rgb=Image.merge("RGB",[img_r,img_g,img_b])
plt.subplot(155)

plt.imshow(img_rgb)
plt.title("img_rgb", fontsize=20)
```

```python
img_rgbre=img_rgb.resize((50,50))
plt.imshow(img_rgbre)
```

```python
img1=img_rgb.transpose(Image.FLIP_LEFT_RIGHT)  #水平翻转
plt.imshow(img1)
```
