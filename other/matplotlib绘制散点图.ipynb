```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#设置黑体为默认字体
plt.rcParams["font.sans-serif"]="SimHei"
#设置正常显示负号
plt.rcParams["axes.unicode_minus"]=False
```

```python
n=100
#生成x坐标
x=np.random.normal(0,1,n)  
y=np.random.normal(0,1,n)

x1=np.random.uniform(-4,4,(1,n))
y1=np.random.uniform(-4,4,(1,n))
```

```python
plt.scatter(x,y,color="red",marker="*",label="正态分布")   #根据坐标绘制点
plt.scatter(x1,y1,color="blue",marker="*",label="均匀分布")
plt.legend()
plt.title("标准正态分布",fontsize=20)
plt.text(2.5,2.5,"均 值:0\n标准差:1",fontsize=15)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel("横坐标",fontsize=18)
plt.ylabel("纵坐标",fontsize=18)
plt.show()
```
