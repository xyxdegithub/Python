```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#要注意 \ 转义字符
train_path = "C:\\Users\\xyx\\.keras\\datasets\\iris_training.csv"
names = ["花萼长度", "花萼宽度", "花瓣长度", "花瓣宽度", "种类"]
pd.read_csv(train_path, header=0, names=names)
```

```python
#将dataframe转为array
datas = np.array(pd.read_csv(train_path, header=0, names=names))
```

```python
#下面画花萼长度和花萼宽度与花种类的图
#花萼长度
sepal_length = datas[:, 0]
print(sepal_length)
```

```python
#花萼宽度
sepal_width = datas[:, 1]
print(sepal_width)
```

```python
#花种类
species = datas[:, 4]
print(species)
print(len(species))
```

```python
#山鸢尾(蓝色),变色鸢尾（红色）,弗吉尼亚鸢尾（绿色）
#设置黑体为默认字体
plt.rcParams["font.sans-serif"] = "SimHei"
#设置正常显示负号
plt.rcParams["axes.unicode_minus"] = False
plt.figure(figsize=(6, 6))
plt.scatter(sepal_length, sepal_width, c=species, cmap="brg")
plt.xlabel(names[0], fontsize=15)
plt.ylabel(names[1], fontsize=15)
plt.title("鸢尾花数据集(山鸢尾-蓝色,变色鸢尾-红色,弗吉尼亚鸢尾-绿色)", fontsize=15)
```

```python
#花瓣长度
petal_length = datas[:, 2]
#花瓣宽度
petal_width = datas[:, 3]
plt.figure(figsize=(6, 6))
plt.scatter(petal_length, petal_width, c=species, cmap="brg")
plt.xlabel(names[2], fontsize=15)
plt.ylabel(names[3], fontsize=15)
plt.title("鸢尾花数据集(山鸢尾-蓝色,变色鸢尾-红色,弗吉尼亚鸢尾-绿色)", fontsize=15)
```

```python
#四种属性两两组合画图
plt.figure(figsize=(20, 20))
plt.suptitle("鸢尾花数据集(山鸢尾-蓝色,变色鸢尾-红色,弗吉尼亚鸢尾-绿色)", y=1,fontsize=20)
for i in range(4):
    for j in range(4):
        plt.subplot(4, 4, 4 * i + (j + 1))
        if(i==j):
            plt.text(0.3,0.4,names[i], fontsize=15)
        else:
            plt.scatter(datas[:,j],datas[:,i],c=species,cmap="brg")
        if(i==0):
            plt.title(names[j], fontsize=15)
        if(j==0):
            plt.ylabel(names[i], fontsize=15)
plt.tight_layout()
```
