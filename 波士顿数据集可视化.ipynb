```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
#设置黑体为默认字体
plt.rcParams["font.sans-serif"] = "SimHei"
#设置正常显示负号
plt.rcParams["axes.unicode_minus"] = False
```

```python
# CRIM：城镇人均犯罪率。
# ZN：住宅用地超过 25000 sq.ft. 的比例。
# INDUS：城镇非零售商用土地的比例。
# CHAS：查理斯河空变量（如果边界是河流，则为1；否则为0）。
# NOX：一氧化氮浓度。
# RM：住宅平均房间数。
# AGE：1940 年之前建成的自用房屋比例。
# DIS：到波士顿五个中心区域的加权距离。
# RAD：辐射性公路的接近指数。
# TAX：每 10000 美元的全值财产税率。
# PTRATIO：城镇师生比例。
# B：1000（Bk-0.63）^ 2，其中 Bk 指代城镇中黑人的比例。
# LSTAT：人口中地位低下者的比例。
# MEDV：自住房的平均房价，以千美元计。
boston_housing=tf.keras.datasets.boston_housing
#把所有数据都当做训练集
(train_x,train_y),(test_x,test_y)=boston_housing.load_data(test_split=0)
```

```python
print(len(train_x))
```

```python
train_xs=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTPATIO","B","LSTAT"]
```

```python
plt.figure(figsize=(20,20)) #设置图的大小尺寸
plt.suptitle("各个特征与房价的关系",y=1,fontsize=30)  #设置全局变量
#绘制每个属性与房价的关系
for i in range(0,13):
    plt.subplot(4,4,(i+1)) #分成16份
    plt.scatter(train_x[:,i],train_y)
    plt.xlabel(train_xs[i])
    plt.ylabel("房价(千美元)")
plt.tight_layout() #使图紧凑
plt.show()
```
